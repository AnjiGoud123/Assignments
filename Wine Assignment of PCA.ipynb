```python
x = df.iloc[:,0:]
```

```python
import scipy.cluster.hierarchy as shc
```

```python
# construction of dendrogram 
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))
plt.title("Customer Dendrogram")
dend = shc.dendrogram(shc.linkage(x, method='complete'))
```

```python
# forming a group using clusters
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean',linkage='complete')
```

```python
y = cluster.fit_predict(x)
y = pd.DataFrame(y)
y.value_counts()
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(x.iloc[:,0],x.iloc[:,1],c=cluster.labels_,cmap='rainbow')
```

```python
# intializing KMeans
from sklearn.cluster import KMeans
KMeans = KMeans(n_clusters=5)
```

```python
# fitting with inputs
KMeans = KMeans.fit(x)
```

```python
#predicting the clusters
y = KMeans.predict(x)

y = pd.DataFrame(y)
y.value_counts()
```

```python
# DBSCAN
from sklearn.preprocessing import StandardScaler
stscaler = StandardScaler().fit(df)
x = stscaler.transform(df)
x
```

```python
from sklearn.cluster import DBSCAN
DBSCAN()
```

```python
dbscan = DBSCAN(eps=0.75, min_samples=3)
dbscan.fit(x)
```

```python
# noisy samples are given the labels as -1
dbscan.labels_
```

```python
c1 = pd.DataFrame(dbscan.labels_,columns=["cluster"])
c1.value_counts()
```

```python
df_new = pd.concat([df,c1],axis=1)
```

```python
noisedata = df_new[df_new['cluster']==-1]
print(noisedata)
finaldata = df_new[df_new['cluster']==0]
finaldata
```
