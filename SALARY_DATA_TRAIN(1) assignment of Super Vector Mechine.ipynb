```python
df.dropna(inplace=True)
df
```

```python
df['Salary'].value_counts()
```

```python
# apply label encoding for all variables
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df["workclass"] = LE.fit_transform(df["workclass"])
df["education"] = LE.fit_transform(df["education"])
df["maritalstatus"] = LE.fit_transform(df["maritalstatus"])
df["occupation"] = LE.fit_transform(df["occupation"])
df["relationship"] = LE.fit_transform(df["relationship"])
df["race"] = LE.fit_transform(df["race"])
df["sex"] = LE.fit_transform(df["sex"])
df["native"] = LE.fit_transform(df["native"])
```

```python
df.head()
```

```python
#  split the data into  x and y
X = df.iloc[:,0:13]
Y = df['Salary']
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X
SS_X = pd.DataFrame(SS_X)
```

```python
# Data partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(SS_X,Y,test_size=0.25,random_state=10)
```

```python
# model fit
from sklearn.svm import SVC
clf = SVC(kernel='linear',C=1.0)
clf.fit(X_train,Y_train)
```

```python
Y_pred_train = clf.predict(X_train)
Y_pred_test= clf.predict(X_test)
```

```python
# metrics 
from sklearn import metrics
print("Training Accuracy:",metrics.accuracy_score(Y_train,Y_pred_train).round(2))
print("Testing Accuracy:",metrics.accuracy_score(Y_test,Y_pred_test).round(2))
```

```python
# model fit
from sklearn.svm import SVC
clf = SVC(kernel='poly',degree=3)
clf.fit(X_train,Y_train)
```

```python
Y_pred_train = clf.predict(X_train)
Y_pred_test= clf.predict(X_test)
```

```python
# metrics 
from sklearn import metrics
print("Training Accuracy:",metrics.accuracy_score(Y_train,Y_pred_train).round(2))
print("Testing Accuracy:",metrics.accuracy_score(Y_test,Y_pred_test).round(2))
```

```python
# model fit
from sklearn.svm import SVC
clf = SVC(kernel='rbf',gamma='scale')
clf.fit(X_train,Y_train)
```
