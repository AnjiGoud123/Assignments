```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()

df['ShelveLoc']=LE.fit_transform(df['ShelveLoc'])
df['Urban']=LE.fit_transform(df['Urban'])
df['US']=LE.fit_transform(df['US'])
df.head()
```

```python
df["Sales"].mean()
```

```python
def f1(x):
    if x >= 7.496325:
        return 0
    else:
        return 1
```

```python
df["Sales_new"] = df["Sales"].apply(f1)
```

```python
df.head()
```

```python
df["Sales_new"]
```

```python
# split the variables into x and y
y = df["Sales_new"]
x = df.iloc[:,1:11]
```

```python
# data partition
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
# model fit
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier(criterion='gini',max_depth=7)
DT.fit(x_train,y_train)
```

```python
y_pred_train = DT.predict(x_train)
y_pred_test = DT.predict(x_test)
```

```python
print('# of nodes:',DT.tree_.node_count)
print("maximum depth:",DT.tree_.max_depth)
```

```python
from sklearn.metrics import accuracy_score
ac1 =  accuracy_score(y_train,y_pred_train)
print("Training Accuracy:",ac1.round(3))
```

```python
ac2 = accuracy_score(y_test,y_pred_test)
print("Test Accuracy:",ac2.round(3))
```
